```
#filtering republican data based on swing states
swing = ['Colorado', 'Florida', 'Iowa', 'Michigan', 'Nevada', 'New Hampshire', 'North Carolina', 'Ohio', 'Pennsylvania', 'Virginia', 'Wisconsin']
swing_df = cnty_fts["state"].isin(swing)
cnty_fts[swing_df].drop(cnty_fts.columns[0], axis=1)
```
